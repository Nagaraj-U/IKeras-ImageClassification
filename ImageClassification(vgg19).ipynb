```
import numpy as np
import keras
import tensorflow as tf
```

#importing libraries


```
from keras.preprocessing.image import load_img,img_to_array
from keras.applications.vgg19 import VGG19,decode_predictions,preprocess_input
from tensorflow.keras.models import Sequential

from IPython.core.display import display
```

#making constant files

```
FILE_1='05 City.jpg'
FILE_2='07 Stairs.jpg'
FILE_3='11 Shoe.jpg'
FILE_4='10 Red Shoes.jpg'
FILE_5='02 Couple.jpg'
FILE_6='08 Doorway.jpg'
```

```
pic=load_img(FILE_1,target_size=(224,224)) #as declared in vgg19 model size is 224/224
display(pic)
```

```
image_array=img_to_array(pic)
image_array.shape
```

#adding additional column

```
expanded_array=np.expand_dims(image_array,axis=0)
expanded_array.shape
```

#preprocessing Input array

```
preprocess=preprocess_input(expanded_array)
```

#loading model with pretrained weights

```
vgg19_model=VGG19(weights='imagenet')
```

#making predictions

```
predictions=vgg19_model.predict(preprocess)
decode_predictions(predictions)
```

#making separate function to perform all above steps

```

def format_image_vgg19(file):
  pic = load_img(file,target_size=(224,224))
  image_array=img_to_array(pic)
  expanded_array=np.expand_dims(image_array,axis=0)
  return preprocess_input(expanded_array)
```

```
data=format_image_vgg19(FILE_5)
predictions=vgg19_model.predict(data)
display(load_img(FILE_5))
decode_predictions(predictions)
```

```
data=format_image_vgg19(FILE_6)
predictions=vgg19_model.predict(data)
display(load_img(FILE_6))
decode_predictions(predictions)
```

```
data=format_image_vgg19(FILE_4)
predictions=vgg19_model.predict(data)
display(load_img(FILE_4))
decode_predictions(predictions)
```
